```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train_data = pd.read_csv('./dataset/train_data.csv')
test_data = pd.read_csv('./dataset/test_data.csv')
```

```python
train_data['HP'] = np.nan
for n in train_data['type'].value_counts().keys():
    
    if n == 0 or n == 4 or n == 5 or n == 6 or n == 7:
        train_data.loc[train_data['type'] == n, 'HP'] = 'HP30'
        test_data.loc[test_data['type'] == n, 'HP'] = 'HP30'
    elif n == 1:
        train_data.loc[train_data['type'] == n, 'HP'] = 'HP20'
        test_data.loc[test_data['type'] == n, 'HP'] = 'HP20'
    elif n == 2:
        train_data.loc[train_data['type'] == n, 'HP'] = 'HP10'
        test_data.loc[test_data['type'] == n, 'HP'] = 'HP10'
    else:
        train_data.loc[train_data['type'] == n, 'HP'] = 'HP50'
        test_data.loc[test_data['type'] == n, 'HP'] = 'HP50'

# train_data['HP'] = train_data['HP'].astype('int64')
# test_data['HP'] = test_data['HP'].astype('int64')

train_data
```

```python
train_data.drop('type', axis=1, inplace=True)
test_data.drop('type', axis=1, inplace=True)
```
