## 4.3 Function Define

### 4.3.1 ì‹œê³„ì—´ ë°ì´í„° ë¶„ì„ì„ ìœ„í•œ ì „ì²˜ë¦¬(for LSTM, CNN)

```python
def temporalize(x, y, timestep):
    output_X = []
    output_y = []
    for i in range(len(x) - timestep + 1 ):
        t = []
        for j in range(timestep):
            t.append(x[[(i + j)], :])
        output_X.append(t)
        output_y.append(y[i + timestep - 1])
    return np.squeeze(np.array(output_X)), np.array(output_y)
```

### 4.3.2 reconstruction error plot

```python
def recon_err_plot(x, x_pred, y, threshold=0):
    mse = np.mean(np.power(x.reshape(x.shape[0], -1) - x_pred.reshape(x_pred.shape[0], -1), 2), axis=1)
    error_df = pd.DataFrame({'Reconstruction_error': mse, 'True_class': y})
    error_df = error_df.reset_index()

    groups = error_df.groupby('True_class')
    fig, ax = plt.subplots(figsize=(15, 8))
    for name, group in groups:
        ax.plot(group.index, group.Reconstruction_error, marker='o', ms=3.5, linestyle='',
                label= "Break" if name == 1 else "Normal")
    ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
    ax.legend()
    plt.title("Reconstruction error plot")
    plt.ylabel("Reconstruction error")
    plt.xlabel("Data point index")
    plt.grid()
    plt.show()

    return error_df
```

### 4.3.3 precision, recall, f1 curve

```python
def prec_rec_f1_curve(y, score, pos = 1) :
    precision, recall, thresholds  = precision_recall_curve(y, score, pos_label=1)
    f1 = 2 / (1/precision + 1/recall)

    plt.figure(figsize = (12,8))

    plt.plot(thresholds, np.delete(precision, -1), label = 'precision')
    plt.plot(thresholds, np.delete(recall, -1), label = 'recall')
    plt.plot(thresholds, np.delete(f1, -1), label = 'f1')
    plt.xlabel('Anomaly Score')
    plt.legend()
    plt.grid()
    plt.show()

    return precision, recall, f1, thresholds
```

```python
def classification_report2(y, pred, thresholds):
    pred_temp = np.where(pred > thresholds , 1, 0)

    print('< confusion matrix >\n')
    print(confusion_matrix(y, pred_temp))
    print('\n' + '='*60 + '\n')

    print('< classification_report >\n')
    print(classification_report(y, pred_temp))

    return confusion_matrix(y, pred_temp)
```

```python
# HP: 10
il_10 = Input(shape=(7, ), name='HP_10')
hl_10 = Dense(16, activation='relu')(il_10)
hl_10 = Dense(8, activation='relu')(hl_10)
hl_10 = Dense(2, activation='relu')(hl_10)
hl_10 = Dense(8, activation='relu')(hl_10)
hl_10 = Dense(16, activation='relu')(hl_10)

# HP: 20
il_20 = Input(shape=(7, ), name='HP_20')
hl_20 = Dense(16, activation='relu')(il_20)
hl_20 = Dense(8, activation='relu')(hl_20)
hl_20 = Dense(2, activation='relu')(hl_20)
hl_20 = Dense(8, activation='relu')(hl_20)
hl_20 = Dense(16, activation='relu')(hl_20)

# HP: 30
il_30 = Input(shape=(7, ), name='HP_30')
hl_30 = Dense(16, activation='relu')(il_30)
hl_30 = Dense(8, activation='relu')(hl_30)
hl_30 = Dense(2, activation='relu')(hl_30)
hl_30 = Dense(8, activation='relu')(hl_30)
hl_30 = Dense(16, activation='relu')(hl_30)

# HP: 50
il_50 = Input(shape=(7, ), name='HP_50')
hl_50 = Dense(16, activation='relu')(il_50)
hl_50 = Dense(8, activation='relu')(hl_50)
hl_50 = Dense(2, activation='relu')(hl_50)
hl_50 = Dense(8, activation='relu')(hl_50)
hl_50 = Dense(16, activation='relu')(hl_50)

cl = Concatenate()([hl_10, hl_20, hl_30, hl_50])
ol = Dense(1, activation='sigmoid')(cl)

model_basic_ae = Model([il_10, il_20, il_30, il_50], ol)

model_basic_ae.compile(loss=keras.losses.binary_crossentropy,
                       metrics=['accuracy'],
                       optimizer='adam')

model_basic_ae.summary()
```
